```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import os
```

```python
df = pd.read_csv(os.getcwd() + '/Desktop/Pokemon.csv', index_col=0)
```

```python
df.head()
```

```python
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
```

```python
sns.boxplot(data=df)
```

```python
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
```

```python
sns.boxplot(data=stats_df)
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Type 1', y='Attack', data=df)
```

```python
stats_df.head()
```

```python
melted_df = pd.melt(stats_df, id_vars=["Name", "Type 1", "Type 2"], var_name="Stats")
melted_df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Stats', y='value', data=melted_df, hue='Type 1', split=True)
plt.ylim(0, 260)
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

```python
g = sns.factorplot(x = 'Type 1', y='Attack', data=df, hue='Stage', col='Stage', kind='swarm')
g.set_xticklabels(rotation=-45)
```
